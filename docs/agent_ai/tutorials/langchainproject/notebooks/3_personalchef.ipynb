```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from langchain.tools import tool
from typing import Dict, Any
from tavily import TavilyClient
from langgraph.checkpoint.memory import InMemorySaver

from dotenv import load_dotenv
load_dotenv()
```

```python
## tool
tavily_client = TavilyClient()

@tool
def web_search(query: str) -> Dict[str, Any]:
    """Search the web for information"""
    return tavily_client.search(query)
```

```python
config={"configurable":{"thread_id":"test-1"}}
```

```python
agent = create_agent(model="groq:qwen/qwen3-32b",
                     tools=[web_search],
                    checkpointer=InMemorySaver(),
)
```

```python
system_message = HumanMessage(content="You are a helpful personal chef assistant.")
user_message = HumanMessage(content="I have chicken, broccoli, and rice. What can I make for dinner it should be different from the previous meal?")
response = agent.invoke(
    {"messages": [system_message, user_message]}, config=config)
```
